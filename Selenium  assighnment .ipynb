# 6
Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 – September 20) from trak.in.

```python
#6
#importing webdriver
driver=webdriver.Chrome("chromedriver.exe")
time.sleep(4)
url="https://trak.in/india-startup-funding-investment-2015/"
driver.get(url)  #getting the funding page
time.sleep(4)

#creating empty boxex
date=[]
s_name=[]
industry=[]
sub=[]
location=[]
i_name=[]
i_type=[]
amount=[]
#getting the data of the month from july to september
for i in range(48,51):
    # Date
    dt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))
    for d in dt:
        date.append(d.text)

    #startupname
    s=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[3]'.format(i))
    for d in s:
        s_name.append(d.text)

    #industry
    a=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[4]'.format(i))
    for d in a:
        industry.append(d.text)
    #sub
    su=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[5]'.format(i))
    for d in su:
        sub.append(d.text)
    #location
    lo=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[6]'.format(i))
    for d in lo:
        location.append(d.text)
    #i_name
    ina=driver.find_elements_by_xpath("//*[@id='tablepress-{}']/tbody/tr/td[7]".format(i))
    for d in ina:
        i_name.append(d.text)
    #i_type
    ty=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[8]'.format(i))
    for d in ty:
        i_type.append(d.text)
    #amount
    am=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[9]'.format(i))
    for d in am:
        amount.append(d.text)

time.sleep(2)
#creating dataframe
ans6=pd.DataFrame({})
ans6["Date"]=date
ans6["Startup_Name"]=s_name
ans6["Industry"]=industry
ans6["Sub"]=sub
ans6["Location"]=location
ans6["Investors name"]=i_name
ans6["Investment_type"]=i_type
ans6["Amount"]=amount
ans6
```

```python
#7
#importing webdriver
driver=webdriver.Chrome("chromedriver.exe")
time.sleep(4)
url="https://www.digit.in/"
driver.get(url)  #getting the website
time.sleep(4)
driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[4]/ul/li[4]").click()  #clicking on top 10 section
time.sleep(2)
driver.find_element_by_xpath("/html/body/div[3]/div/div/div[2]/div[5]/div[1]/div/button[2]").click() #clicking on laptop catogery
time.sleep(2)
driver.find_element_by_xpath("/html/body/div[3]/div/div/div[2]/div[5]/div[2]/div[3]/a/div[2]/p").click() #best gaming laptops in india
time.sleep(2)
#getting urls of all the laptop available
urls=[]
url=driver.find_elements_by_xpath("//span[@class='datahreflink spec']")
for i in url:
    urls.append(i.get_attribute("data-href"))
#creating empty boxex
brand_name=[]
product_name=[]
os=[]
ram=[]
storaget=[]
storagec=[]
ram_type=[]
weight=[]
ds=[]
dr=[]
processor=[]
p_core=[]
dim=[]
clock=[]
gpu=[]
cashe=[]
boost=[]
       
#extracting data
for i in urls[:5]:
    driver.get(i)
    time.sleep(4)
    #extracting brand name
    try:
        b=driver.find_element_by_xpath('//div[@class="heading-wraper"]')
        brand_name.append(b.text)
    except NoSuchElementException:
        #b=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        brand_name.append("--")
    brand_name = [x.split()[0] for x in brand_name]
   #Product name
    try:
        p=driver.find_element_by_xpath("//*[@id='specs-basic-information']/div[2]/table/tbody/tr[1]/td[3]")
        product_name.append(p.text)
    except NoSuchElementException:
        product_name.append("--")
   #os
    try:
        r=driver.find_element_by_xpath("//*[@id='specs-basic-information']/div[2]/table/tbody/tr[2]/td[3]")
        os.append(r.text)
    except NoSuchElementException:
        os.append("--")
   #ram
    try: 
        nr=driver.find_element_by_xpath('//*[@id="specs-memory"]/div[2]/table/tbody/tr[1]/td[3]')
        ram.append(nr.text)
    except NoSuchElementException:
        ram.append("--")
   #Storage Type
    try:
        re=driver.find_element_by_xpath("//*[@id='specs-storage']/div[2]/table/tbody/tr[1]/td[3]")
        storaget.append(re.text)
    except NoSuchElementException:
        storaget.append("--")
   #Storage capacity
    try:
        re=driver.find_element_by_xpath("//*[@id='specs-storage']/div[2]/table/tbody/tr[2]/td[3]")
        storagec.append(re.text)
    except NoSuchElementException:
        storagec.append("--") 
   #Ram type
    try:
        ed=driver.find_element_by_xpath("//*[@id='specs-memory']/div[2]/table/tbody/tr[2]/td[3]")
        ram_type.append(ed.text)
    except NoSuchElementException:
        ram_type.append("--")
   #weight
    try:
        ed=driver.find_element_by_xpath("//*[@id='specs-physical-specifications']/div[2]/table/tbody/tr[1]/td[3]")
        weight.append(ed.text)
    except NoSuchElementException:
        weight.append("--")
   #Display size
    try:
        ed=driver.find_element_by_xpath("//*[@id='specs-display']/div[2]/table/tbody/tr[2]/td[3]")
        ds.append(ed.text)
    except NoSuchElementException:
        ds.append("--")
   #Display resolution
    try:
        ed=driver.find_element_by_xpath("//*[@id='specs-display']/div[2]/table/tbody/tr[1]/td[3]")
        dr.append(ed.text)
    except NoSuchElementException:
        dr.append("--")
   #Processor
    try:
        odd=driver.find_element_by_xpath("//*[@id='specs-processor']/div[2]/table/tbody/tr[1]/td[3]")
        processor.append(odd.text)
    except NoSuchElementException:
        processor.append("--")
   #Clock speed
    try:
        odd=driver.find_element_by_xpath("//*[@id='specs-processor']/div[2]/table/tbody/tr[2]/td[3]")
        clock.append(odd.text)
    except NoSuchElementException:
        clock.append("--")     
   #p_core
    try:
        ed=driver.find_element_by_xpath("//*[@id='specs-processor']/div[2]/table/tbody/tr[3]/td[3]")
        p_core.append(ed.text)
    except NoSuchElementException:
        p_core.append("--")
   #GPU
    try:
        ed=driver.find_element_by_xpath("//*[@id='specs-processor']/div[2]/table/tbody/tr[4]/td[3]")
        gpu.append(ed.text)
    except NoSuchElementException:
        gpu.append("--") 
   #Cache
    try:
        ed=driver.find_element_by_xpath("//*[@id='specs-processor']/div[2]/table/tbody/tr[5]/td[3]")
        cashe.append(ed.text)
    except NoSuchElementException:
        cashe.append("--")  
   #Boost clock speed
    try:
        ed=driver.find_element_by_xpath("//*[@id='specs-processor']/div[2]/table/tbody/tr[6]/td[3]")
        boost.append(ed.text)
    except NoSuchElementException:
        boost.append("--")   
   #Laptop dimension
    try:
        ed=driver.find_element_by_xpath("//*[@id='specs-physical-specifications']/div[2]/table/tbody/tr[2]/td[3]")
        dim.append(ed.text)
    except NoSuchElementException:
        dim.append("--")


time.sleep(2)
ans7=pd.DataFrame({})
#creating dataframe
ans7["Brand_Name"]=brand_name
ans7["Product_Name"]=product_name
ans7["Operating system"]=os
ans7["Ram"]=ram
ans7["Ram_Type"]=ram_type
ans7["Storage_Type"]=storaget
ans7["Storage_Capacity"]=storagec
ans7["Weight(in KG)"]=weight
ans7["Dimensions(In Mm)"]=dim
ans7["Display_size"]=ds
ans7["Display_resolution"]=dr
ans7["Processor"]=processor
ans7["processor_core"]=p_core
ans7["GPU"]=gpu
ans7["Cache L3"]=cashe
ans7["Clock Speed"]=clock
ans7["Boost clock speed"]=boost
ans7
```
